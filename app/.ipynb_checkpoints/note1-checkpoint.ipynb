```python
students=pd.read_csv("static/CSV/students.csv",sep=',', delimiter=None, header='infer', encoding='cp1252',engine='python')
```

```python
sg_data=students[["GENDER","AC_YEAR","AC_GRADE","Physics","STD","PASSED","FAILED","NUM_SUBJ","MEAN_CRED","ENROL_ECTS","PASS_ECTS","ALPHA"]]
```

```python
df=sg_data
df=df[["GENDER","AC_YEAR",(df.columns[3])]]
df=df.dropna(axis=0, how='any')
s = df.groupby(["GENDER","AC_YEAR"]).size()
df["NUM_STUD"] = df.set_index(["GENDER","AC_YEAR"]).index.map(s.get)
df=df[["AC_YEAR","GENDER","NUM_STUD"]]
df.drop_duplicates(subset=["GENDER","AC_YEAR"], keep='first', inplace=True)
df=df.pivot_table(index="AC_YEAR",columns="GENDER",values="NUM_STUD")
df=df.reset_index()
```
