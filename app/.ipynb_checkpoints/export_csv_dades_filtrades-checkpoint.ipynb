```python
from flask import Flask, render_template
import pandas as pd

app = Flask(__name__)

file1="static/CSV/Dadespersnomespreinsdec16originalescoles.csv"
file2="static/CSV/Dadesqualiffaseinicialdec16original.csv"

def open_registered(file1):
    
    registered=pd.read_csv(file1,sep=',', delimiter=None, header='infer',engine='python',dtype={"CODI_EXPEDIENT":str, "VIA_ACCES":str,"TIPUS_ACCES":str,"CP_FAMILIAR":str,"CP_CENTRE_SEC":str})
    
    #Students from GETI
    registered=registered[registered["CODI_PROGRAMA"]==752]
    registered.drop(labels=["CODI_PROGRAMA"],axis=1,inplace=True)
    #Students registered in 2010 or later
    registered=registered[registered["ANY_ACCES"]>=2010]
    #Access via: PAU
    registered=registered[registered["VIA_ACCES"]=="0"]
    registered.drop(labels=["VIA_ACCES"],axis=1,inplace=True)
    registered=registered[registered["TIPUS_ACCES"]=="1"]
    registered.drop(labels=["TIPUS_ACCES"],axis=1,inplace=True)
    #With no info
    registered=registered.dropna(axis=0, how='any')
    
    #Name columns
    registered.columns=["EXPEDIENT_CODE","GENDER","FAM_CODE","AC_YEAR","AC_GRADE","HSCHOOL","HS_CODE"]
    
    #Delete duplicates (Keep first time registered at ETSEIB)
    registered=registered.sort_values(by=["EXPEDIENT_CODE","AC_YEAR","AC_GRADE"],ascending=True)
    registered.drop_duplicates(subset="EXPEDIENT_CODE", keep='first', inplace=True)
    
    return registered

def open_grades(file2):
    grades=pd.read_csv(file2,sep=',', delimiter=None, header='infer',engine='python',dtype={"CODI_EXPEDIENT":str, "CODI_UPC_UD":str})
    
    #Students from GETI
    grades=grades[grades["CODI_PROGRAMA"]==752]
    grades.drop(labels=["CODI_PROGRAMA"],axis=1,inplace=True)
    #Courses taken in 2010 or later
    grades=grades[grades["CURS"]>=2010]
    #Drop unnecessary info
    grades.drop(labels=["CREDITS","NOTA_PROF","NOTA_NUM_AVAL","GRUP_CLASSE"],axis=1,inplace=True)
    #With no info
    grades=grades.dropna(axis=0, how='any')
    
    #Name columns
    grades.columns=["EXPEDIENT_CODE","COURSE_CODE","YEAR","QUARTER","PASS","GRADE"]
    
    #Number of attempts per course
    grades["ATTEMPTS"]=None
    s = grades.groupby(["EXPEDIENT_CODE", "COURSE_CODE"]).size()
    grades["ATTEMPTS"] = grades.set_index(["EXPEDIENT_CODE","COURSE_CODE"]).index.map(s.get)
    
    grades=grades.sort_values(by=["EXPEDIENT_CODE","COURSE_CODE","YEAR","QUARTER"],ascending=True)
    
    return grades

def one_in_another(file1, file2):
    
    #Call this function to get the 2 Dataframes
    registered=open_registered(file1)
    grades=open_grades(file2)
    
    registered=registered[registered["EXPEDIENT_CODE"].isin(grades["EXPEDIENT_CODE"])]
    grades=grades[grades["EXPEDIENT_CODE"].isin(registered["EXPEDIENT_CODE"])]
    
    return registered, grades

registered,grades=one_in_another(file1, file2)
hs=registered["HSCHOOL"].unique()

@app.route("/")
def index():
    hsch=["a","b","c"]
    return render_template("trying.html",hsch=hsch)

if __name__ == '__main__':
    from werkzeug.serving import run_simple
    run_simple('localhost', 9000, app)
```
